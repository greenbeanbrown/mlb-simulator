```python
def scrape(year):
    
    #url, pages = seasons_dict
    
    # This will hold the entire seasons data (agg of dfs)
    agg_df = pd.DataFrame()
    
    # Initialize web driver object and send GET request to grab HTML 
    driver = webdriver.Firefox()
    #driver.get(url)

    # Accept cookies to move forward
    #WebDriverWait(driver, 60).until(EC.visibility_of_element_located((By.ID, "onetrust-accept-btn-handler")))
    #cookies_button = driver.find_element_by_id('onetrust-accept-btn-handler')
    #cookies_button.click()
    
    # Parse out the number of pages
    #WebDriverWait(driver, 60).until(EC.visibility_of_element_located((By.CLASS_NAME, "stats-table-pagination__info")))
    #raw_pages_text = driver.find_element_by_class_name('stats-table-pagination__info').text
    #raw_pages_text = raw_pages_text.replace('\n','')
    #pages = int(raw_pages_text[raw_pages_text.find('of')+3:])
    
    # Loop through each page and scrape data
    for page_nbr in range(1, 99):
        
        # Create url
        url = 'https://www.mlb.com/stats/{}?page={}'.format(str(year), str(page_nbr))
        driver.get(url)
        
        # Accept cookies to move forward
        if page_nbr == 1:
            WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.ID, "onetrust-accept-btn-handler")))
            cookies_button = driver.find_element_by_id('onetrust-accept-btn-handler')
            cookies_button.click()
    
        # Check if there is no data and end the loop if so
        data_msg = driver.find_elements_by_class_name('no-results-secondary-text-1dzDTxyH')
        if len(data_msg) > 0:
            break
        else:
        
            # Scrape data, need to use BeautifulSoup to get JS rendered tables
            soup = BeautifulSoup(driver.page_source, 'lxml')
            tables = soup.find_all('table')
            # The player stats are the 1st table
            df = pd.read_html(str(tables))[0]
            agg_df = agg_df.append(df)
            
    return agg_df
```

```python
year = 2021

scrape(year)
```
